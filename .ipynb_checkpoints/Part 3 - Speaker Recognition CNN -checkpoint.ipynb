```python
#Function from Alex Lin - GA TA or GAHTAH .... jk
def bash(string):
    '''run a bash command and return stdout lines as a py list'''
    stdout=(subprocess.Popen(string, shell=True, stdout=subprocess.PIPE).stdout.read())
    output = str(stdout)[2:-1].split('\\n')[:-1]
    #output = str(stdout).split('\\n')[1:-1]
    #output[0] = output[0][2:]
    return output
```

### Load in Data somehow 

```python
#Get path to each folder
path_to_images = 'E:/spectrogram/'
list_of_folders = bash('ls ' + path_to_images)
list_of_folders_abs = [(path_to_images + folder +'/') for folder in list_of_folders]
```

```python
#Get path to each spectrogram
spectrograms_abs = [] 
for spectro_folder in list_of_folders_abs:
    spectrograms = bash('ls ' + spectro_folder)
    for spectrogram in spectrograms:
        spectrogram_abs = spectro_folder + spectrogram
        spectrograms_abs.append(spectrogram_abs)
```

```python
# Store absolute spectrogram file paths in a csv
with open('abs_spectrogram_path.txt', 'w') as f:
    for item in spectrograms_abs:
        f.write("%s\n" % item)
```

```python
# Read absolute spectrogram file paths from a csv
abs_spectrogram_path = []
with open('abs_spectrogram_path.txt', 'r') as f:
        list_ = csv.reader(f)
        for row in list_:
            abs_spectrogram_path.append(row[0])
            
```
