```python
# Get all the spectrograms in the first six files
# then put in a numpy array
# Get labels too

img_array = []
labels = []

for path in abs_spectrogram_path:
    if path == 'E:/spectrogram/06_mp3/06_mp3-314.png':
        img = cv2.imread(path,1)
        img_norm = img/255
        img_array.append(img)
        labels.append(path[15:21])
        break;
    img = cv2.imread(path,1)
    img_norm = img/255
    img_array.append(img)
    labels.append(path[15:21])
```

```python
# Get np array of data
img_array_nd = np.array(img_array)
```

```python
# input array shape
img_array_nd.shape
```

```python
# label size
labels
```

```python
# Get labels
le = LabelEncoder()
y_cat = le.fit_transform(labels)
y_cat = y_cat.reshape(-1,1)
```

```python
# One hot encode
ohe = OneHotEncoder()
y_cat = ohe.fit_transform(y_cat).toarray()
```

```python
# One hot encoded data
y_cat
```

```python
X = img_array_nd
y = y_cat
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,stratify = y)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

 ## Model from class to try 

```python
model = Sequential()

model.add(Conv2D(filters = 8,                   # number of filters
                         kernel_size = 3,        # height/width of filter
                         padding = "same",
                         activation='relu',      # activation function 
                         input_shape=(513,800,3))) # shape of input (dimensions of training image)

model.add(Conv2D(16, kernel_size = 3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
 
model.add(Conv2D(16, kernel_size = 3, activation='relu'))
model.add(Conv2D(16, kernel_size = 3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Activation('relu'))
# model.add(Dense(128, activation='relu'))
# model.add(Dropout(0.5))
model.add(Dense(6, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, 
          batch_size=32, epochs=2, verbose=1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
labels = model.metrics_names
```

```python
print(str(labels[0]) + ": " + str(score[0]))
print(str(labels[1]) + ": " + str(score[1]))
```
