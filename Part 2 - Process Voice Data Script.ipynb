```python
#Function from Alex Lin
def bash(string):
    '''run a bash command and return stdout lines as a py list'''
    stdout=(subprocess.Popen(string, shell=True, stdout=subprocess.PIPE).stdout.read())
    output = str(stdout)[2:-1].split('\\n')[:-1]
    #output = str(stdout).split('\\n')[1:-1]
    #output[0] = output[0][2:]
    return output
```

## Converting files to 45 min

```python
# Example code: "ffmpeg -i {a}/combined{b}_MP3WRAP.mp3 -ss 0 -t 2700 -acodec copy {a}/combined{b}_45m.mp3;"

base = 'E:/combined/'
files = bash('ls '+base)
destination = 'E:/converted/'

for file in files:
    
    file_conv = file[0:6] + "_converted_45m.mp3"
    
    source = base + file
    #print(source)
    
    destination_file = destination + file_conv
    #print(destination_file)
    
    cmd = "ffmpeg -i " + source + " -ss 0 -t 2700 -acodec copy " + destination_file
    print(cmd)
    
    continue_ = 'y'#input("Do you want to continue converting?")
    
    if continue_ == "y":
        bash(cmd)
    else:
        break;
        
    #bash("ffmpeg -i {a}/combined{b}_MP3WRAP.mp3 -ss 0 -t 2700 -acodec copy {a}/combined{b}_45m.mp3;"")
```

## Converting files to .wav files

```python
# Example code: ffmpeg -i {a}/combined_45m/combined{b}_45m.mp3 -acodec pcm_s16le -ac 1 -ar 16000 {a}/combined_45m/combined{b}_45m.wav;

# -pcm_s16le is codec 16 bit conversion
# -ar is sampling rate (16000samples/sec)
# -ac no. of audio channel

base = 'E:/converted/'
files = bash('ls '+base)
destination = 'E:/wav/'

for file in files:
    
    file_conv = file[0:6] + "_wav_45m.wav"
    
    source = base + file
    #print(source)
    
    destination_file = destination + file_conv
    #print(destination_file)
    
    if os.path.isfile(destination_file): 
        print(destination_file," exists :D , moving on to next file...")
        continue 
        
    
    #"ffmpeg -i source -acodec pcm_s16le -ac 1 -ar 16000 destination_file;"
    cmd = "ffmpeg -i " + source + " -acodec pcm_s16le -ac 1 -ar 16000 "+ destination_file
    print(cmd)
    
    continue_ = 'y'#input("Do you want to continue converting?")
    
    if continue_ == "y":
        os.system(cmd)
    else:
        break;
```

## Removing silence longer than 0.5 seconds 

<p style="font-family:garamond;font-size:120%;"><b>
    <a href=https://digitalcardboard.com/blog/2009/08/25/the-sox-of-silence/>References</a>: https://digitalcardboard.com/blog/2009/08/25/the-sox-of-silence/

```python
# Example code: "sox {a}/combined_45m/combined{b}_45m.wav {a}/combined_45m/combined{b}_45m_fixed.wav silence 1 0.1 1% -1 0.5 1%;"

base = 'E:/wav/'
files = bash('ls '+base)
destination = 'E:/no_silence/'

for file in files:
    
    file_conv = file[0:6] + "_wav_no_sil.wav"
    
    source = base + file
    #print(source)
    
    destination_file = destination + file_conv
    #print(destination_file)
    
    if os.path.isfile(destination_file): 
        print(destination_file," exists :D , moving on to next file...")
        continue 
        
    #"sox {a}/combined_45m/combined{b}_45m.wav {a}/combined_45m/combined{b}_45m_fixed.wav silence 1 0.1 1% -1 0.5 1%;"
    cmd = "sox "+ source +" "+ destination_file +" silence 1 0.1 1% -1 0.5 1%;"
    print(cmd)
    
    continue_ = 'y' #input("Do you want to continue converting?")
    
    if continue_ == "y":
        os.system(cmd)
    else:
        break;
```

# Split file into 5 second segments

```python
# Example code: "ffmpeg -i {a}/combined_45m/combined{b}_45m_fixed.wav -f segment -segment_time 5 -c copy {a}/wave_seg_5sec_45m/out%03d.wav;"

#  In the split folder:
#  Make folders for each audio file
#  Put 5 second segments in those folder

base = 'E:/no_silence/'
files = bash('ls '+base)
destination = 'E:/split/'

for file in files:
    
    file_conv = file[0:6] + "-%03d.wav"
    
    source = base + file
    #print(source)
    
    destination_folder = destination + file[0:6] +"/"
    #print(destination_file)
    
    if os.path.isdir(destination_folder): 
        print(destination_folder," exists :D , moving on to next file...")
        continue 
    else :
        os.mkdir(destination_folder)
        
    #"ffmpeg -i {a}/combined_45m/combined{b}_45m_fixed.wav -f segment -segment_time 5 -c copy {a}/wave_seg_5sec_45m/out%03d.wav;"
    cmd = "ffmpeg -i "+ source +" -f segment -segment_time 5 -c copy "+ destination_folder + file_conv
    print(cmd)
    
    continue_ = 'y'#input("Do you want to continue converting?")
    
    if continue_ == "y":
        os.system(cmd)
    else:
        break;
```

## Convert split audio files to spectrograms

```python
#Example code : os.system("sox {wav} -n spectrogram -r -o {a}/specs_5sec_45m/out{c}.png".format(a=path, b=num, wav=wav_seg, c=cnt))

# The split folder has folders that each correspond to a wav file
# in each folder are the split segments of that file
# 
# Go to each one the segment folders and convert each split wav file to a spectrogram
# Store each spectrogram in, for example, spectrogram/01_mp3/ 
# Then spectrogram/01_mp3

base = 'E:/split/'
folders = bash('ls '+base)
destination = 'E:/spectrogram/'

for folder in folders:
    # source + folder name
    source_of_wavs = base + folder
    
    # destination + same folder name
    destination_of_spectrograms = destination + folder
    
    if os.path.isdir(destination_of_spectrograms): 
        print(destination_of_spectrograms," exists :D , moving on to next folder...")
        continue 
    else :
        continue_ = 'y'#input("Do you want to continue converting?")
        if continue_ == "y":
            os.mkdir(destination_of_spectrograms)
            print(folder, " created")
        else:
            break;
        
    
    #Get list of .wav files in the current folder
    files =  bash('ls '+ source_of_wavs)
    # START TIME
    start = time.time()
    
     
    
    #iterate over each file to convert
    for file in files:
        file_conv = file[0:10] + ".png"
        #print(file_conv)
        
        # Name of WAV file
        wav_file = source_of_wavs + "/" + file
        
        # Name of PNG spectrogram file
        spectrogram_file = destination_of_spectrograms +'/'+ file_conv
        
        #os.system("sox {wav} -n spectrogram -r -o {a}/specs_5sec_45m/out{c}.png".format(a=path, b=num, wav=wav_seg, c=cnt))
        cmd = "sox "+ wav_file +" -n spectrogram -r -o "+ spectrogram_file
        
        os.system(cmd)
    # END TIME
    
    print(folder," ", math.floor(time.time() - start), " seconds to convert to spectrograms")    
       
    
```
