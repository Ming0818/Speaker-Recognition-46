```python
#Function from Alex Lin - GA TA or GAHTAH .... jk
def bash(string):
    '''run a bash command and return stdout lines as a py list'''
    stdout=(subprocess.Popen(string, shell=True, stdout=subprocess.PIPE).stdout.read())
    output = str(stdout)[2:-1].split('\\n')[:-1]
    #output = str(stdout).split('\\n')[1:-1]
    #output[0] = output[0][2:]
    return output
```

## Checking out spectrograms in SciPy

I already have spectrograms outputted from the sox command line tool, I'm just seeing what it looks like in SciPy

#### Get audio path

```python
split = 'E:/split/'

folder_num = 0
file_num = 0

split_folders = bash('ls ' + split)
split_folders_num = split_folders[folder_num]

split_files = bash('ls '+ split + split_folders_num)
split_files_wav = split_files[file_num]

split_folders_num, split_files_wav

abs_path = split + split_folders_num + '/' + split_files_wav
abs_path
```

#### Make a spectrogram

```python
# Gets sample_rate and samples from wav file
sample_rate, samples = wavfile.read(abs_path) 
print('Audio file: ',abs_path ,'\nSample rate: ', sample_rate, ' \n# of samples: ', len(samples))
```

```python
# Creates and displays spectrogram
# This 
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)

# log base e
plt.pcolormesh(times, frequencies, np.log(spectrogram))

# plt.imshow(spectrogram)
# plt.title("First audio file spectrogram")
# plt.ylabel('Frequency [Hz]')
# plt.xlabel('Time [sec]')
#plt.xticks([])
#plt.yticks([])
plt.savefig('foo.png', bbox_inches='tight')
plt.show()


# Example of first audio file
print("First audio file")
audio_path = abs_path
Audio(data = audio_path, autoplay = False, embed=True)
```
